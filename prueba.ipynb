<h4>Conexion de APIREST con flask</h4>


```python
from flask import Flask, jsonify, request

app = Flask(__name__)

# Lista temporal para almacenar las historias clínicas
historias_clinicas = []

# Endpoint para crear una nueva historia clínica
@app.route('/historias_clinicas', methods=['POST'])
def crear_historia_clinica():
    historia_clinica = request.get_json()
    historias_clinicas.append(historia_clinica)
    return jsonify(historia_clinica), 201

# Endpoint para obtener todas las historias clínicas
@app.route('/historias_clinicas', methods=['GET'])
def obtener_historias_clinicas():
    return jsonify(historias_clinicas), 200

# Endpoint para obtener una historia clínica por su ID
@app.route('/historias_clinicas/<int:id>', methods=['GET'])
def obtener_historia_clinica_por_id(id):
    for historia_clinica in historias_clinicas:
        if historia_clinica['id'] == id:
            return jsonify(historia_clinica), 200
    return 'Historia clínica no encontrada', 404

# Endpoint para actualizar una historia clínica por su ID
@app.route('/historias_clinicas/<int:id>', methods=['PUT'])
def actualizar_historia_clinica(id):
    for historia_clinica in historias_clinicas:
        if historia_clinica['id'] == id:
            historia_clinica['diagnostico'] = request.get_json()['diagnostico']
            historia_clinica['tratamiento'] = request.get_json()['tratamiento']
            return jsonify(historia_clinica), 200
    return 'Historia clínica no encontrada', 404

# Endpoint para eliminar una historia clínica por su ID
@app.route('/historias_clinicas/<int:id>', methods=['DELETE'])
def eliminar_historia_clinica(id):
    for historia_clinica in historias_clinicas:
        if historia_clinica['id'] == id:
            historias_clinicas.remove(historia_clinica)
            return 'Historia clínica eliminada', 200
    return 'Historia clínica no encontrada', 404

if __name__ == '__main__':
    app.run(debug=True)
```
